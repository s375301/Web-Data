```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.saramin.co.kr/zf_user/search?search_area=main&search_done=y&search_optional_item=n&searchType=search&searchword=%EB%8D%B0%EC%9D%B4%ED%84%B0%EB%B6%84%EC%84%9D"
url
```

```python
res = requests.get(url)
res
```

```python
res.text
```

```python
my_soup = BeautifulSoup(res.text, 'html.parser')
my_soup
```

```python
my_items = my_soup.find("div", {"class": "content"}).find_all("div",{"class":"item_recruit"})
[item.find("h2",{"class": "job_tit"}).find("a", {"class": "data_layer"}).find("span").text for item in my_items]
```

```python
# title, job_career, company_name, hyperlink under <div class="area_job">
my_jobs = []
for item in my_items:
    job = item.find("div", {"class": "area_job"})
    title = job.find("h2", {"class": "job_tit"}).find("a", {"class": "data_layer"}).find("span").text # tag: find, find_all
    link = "https://www.saramin.co.kr" + job.find("h2", {"class": "job_tit"}).find("a", {"class": "data_layer"})['href'] # props: dict-wise
    career = job.find("div", {"class": "job_condition"}).find_all("span")[1].text # 2nd <span>
    company_name = item.find("div", {"class": "area_corp"}).find("strong", {"class": "corp_name"}).find("a").find("span").text
    my_jobs.append({
        "title": title,
        "link": link,
        "career": career,
        "company": company_name
    })

my_jobs
```

```python
# career constraint
my_jobs = []
for item in my_items:
    job = item.find("div", {"class": "area_job"})
    title = job.find("h2", {"class": "job_tit"}).find("a", {"class": "data_layer"}).find("span").text
    link = "https://www.saramin.co.kr" + job.find("h2", {"class": "job_tit"}).find("a", {"class": "data_layer"})['href']
    career = job.find("div", {"class": "job_condition"}).find_all("span")[1].text

    if "신입" not in career:
        continue

    company_name = item.find("div", {"class": "area_corp"}).find("strong", {"class": "corp_name"}).find("a").find("span").text
    my_jobs.append({
        "title": title,
        "link": link,
        "career": career,
        "company": company_name
    })

my_jobs
```
