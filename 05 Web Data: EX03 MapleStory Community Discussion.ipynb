```python
def maplestory_community(page):
    headers= {'Connection': 'keep-alive'}
    params = {'page': page}
    res = requests.get(url, headers=headers, params=params)
    soup = BeautifulSoup(res.text, 'html.parser')

    posts = []
    for post in my_soup.find('div', {'class': 'notice'}).find_all('li'):
        if post.find('p', {'class': 'list_title'}) is None:
            continue

        title = post.find('p', {'class': 'list_title'}).find('span', {'class': 'title2'}).text
        user = post.find('span', {'class': 'user_name'}).find('a').text
        count = post.find('li', {'class': 'eye_cnt'}).text
        posts.append({
        "title": title,
        "user": user,
        "count": count
        })
    
    return posts
```

```python
all_posts = []
for i in range(1, 10):
    all_posts += maplestory_community(i)

discussions = pd.DataFrame(all_posts)
```
