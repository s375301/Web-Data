## 합(총합)

- sum(list) : sum 함수는 list를 입력값으로 받는다.

```python
# list => [1, 2, 3, 4]
sum([1, 2, 3, 4]) # == 1 + 2 + 3 + 4
```

```python
# 초기값을 부여할 수 있다.
sum([1, 2, 3, 4], 100) # == 100 + 1 + 2 + 3 + 4
```

## 2진수, 8진수, 그리고 16진수 다루기

- bin
- oct
- hex

(참고) 문자열을 반환하는 이유는 해당 수가 "몇진수" 인지 불명확하기 때문

```python
bin(75) # 진수 표현과 함께 문자열로 표현된다.
```

```python
0b1001011
```

```python
oct(75) # 0o
```

```python
hex(75) #0x
```

### 2진수, 8진수, 16진수 => 10진수로 바꾸는 내장함수

- int

```python
int('0b1001011', base=2) # base=2 : 2진수
```

```python
int('1001011', base=2) # base=2 때문에 0b를 제외하여도 된다.
```

```python
int('0o113', base=8)
```

```python
int('113', base=8)
```

```python
int('0x4b', base=16)
```

```python
int('4b', base=16)
```

```python
int('1001011') # base를 지정해주지 않는 경우 default가 10진수이므로 그래도 표현
```

```python
int('1001011', base=10) # == int('1001011')
```

## 수의 나열

- range()
  - range(n) : 0 ~ (n - 1) # range(end)
  - range(start, end) : start ~ (end - 1)
  - range(start, end, step) : start, (start + step), ..., (end - 1)

```python
list(range(10)) # 주의 : 10 - 1
```

```python
list(range(3, 10)) # 주의 : 10 - 1
```

```python
list(range(3, 10, 2)) # 주의 : 10 - 1
```

```python
# 0 부터 10 까지 짝수
list(range(0, 10, 2))
```

```python
# 0 부터 10 까지 홀수
list(range(1, 10, 2))
```
