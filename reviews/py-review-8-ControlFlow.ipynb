### for문의 문법

- for 변수이름 in 이터러블

    - (참고) 이터러블 객체는 list()로 펼쳐볼 수 있다.
  - 변수이름: 꺼내온 값을 저장한 공간

```python
for num in range(3):
    print(num)
```

```python
for c in "Python":
    print(c)
```

```python
for num in [1, 2, 3, 4, 5]: # 리스트 또한 값을 하나씩 꺼내 쓸 수 있는 이터러블 객체
    print(num)
```

```python
my_dict = {
    "a": 100,
    "b": 200,
    "c": 300,
    "d": 400
}
```

```python
"a" in my_dict.keys() # "a" in my_dict 지양
```

```python
100 in my_dict.values()
```

```python
-100 in my_dict.values()
```

```python
for key in my_dict: # 지양
    print(key)
```

```python
for key in my_dict.keys():
    print(key)
```

```python
for value in my_dict.values():
    print(value)
```

```python
for a in my_dict.items():
    print(a) # a는 (key, value)꼴의 튜플 반호나
```

```python
# 튜플의 성질인 "패턴매칭"을 이용에 각각의 값에 저장할 수 있다.
for k, v in my_dict.items():
    print(k, v)
```

```python
# 물론 하나로 받아서 인덱싱을 활용할 수 있지만 위의 방식이 가독성이 높다.
for a in my_dict.items():
    print(a[0], a[1])
```

```python
for c in "Python": # 문자열 역시 하나의 이터러블
    print(c)
```

```python
# enumerate(): 이터러블 값의 순서에도 관심이 있는 경우
for c in enumerate("Python"):
    print(c)
```

```python
# enumerate()도 튜플을 반환하기 때문에 패턴 매칭을 활용할 수 있다.
for i, c in enumerate("Python"):
    if i == 2: # 인덱스 2인 세번째 값 추출
        print(c)
```

```python
for i, num in enumerate([1, 3, 5, 7, 9]):
    print(i, num)
```

```python
# 일반적인 활용
my_list = [1, 3, 5, 7, 9]
for i, num in enumerate(my_list):
    print(i, num)
```

```python
# 홀수만 출력
for i, num in enumerate(my_list):
    if i % 2 == 1:
        print(i, num)
```

```python
# 흔한 반복문 : 횟수가 정해진 반복문
for n in range(5):
    print("Hello")
```

```python
for n in range(1, 5):
    print(f"{n}: Hello")
```

```python
for n in range(1, 10, 2):
    print(n)
```

## 반복문의 키워드: continue, break

- continue: 바로 다음 반복문의 코드를 수행한다.
- break: 해당 반복문 중지

```python
for i in range(5):
    if i == 2:
        continue # i == 2 조건에서, 아래의 코드를 실행하지 않고 i = 3 으로 넘어간다.
    # print(i)
    print(f"{i}: Hello")
```

```python
n = 0
for i in range(5):
    print(n)
    n += 1
    if n == 3:
        break # n == 3 조건에서 반복문을 중단하고 그 다음 코드 진행한다.

print("After Iteration.")
```
