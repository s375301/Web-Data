## 몫과 나머지 연산

- 몫 (//)
- 나머지 (%)
- divmod : 몫과 나머지는 자주 사용되기 때문에 이 같은 내장함수를 지원한다.

```python
17 / 3
```

```python
17 // 3
```

```python
17 % 3
```

```python
divmod(17, 3) # 몫과 나머지를 한번에 구할 수 있다.
```

```python
# divmod 가 튜플을 반환한다는 성질을 이용해서 다음과 같이 이용할 수 있다.
num1, num2 = divmod(17, 3)
num1, num2
```

```python
# 같은 표현
(num1, num2) = divmod(17, 3)
print(num1, num2)
```

# _ 변수 (언더바 변수)
- 마지막 값을 저장하는데 사용
- 결과를 사용하지 않을 때

```python
width = 10
width
```

```python
_ # == width, the last variable
```

```python
tax = 12.5 / 100
price = 100.50
price * tax
```

```python
# _ 사용법 (1) : 마지막 값을 저장하는데 사용
price + _ # total == price*(1+tax)
```

```python
# _ 사용법 (2) : 결과를 사용하지 않는 변수
a, _ = divmod(17, 3)
a
```

```python
def my_fn(a, b, c, d) :
    return d, c, b, a
```

```python
my_fn(1, 2, 3, 4)
```

```python
x, _, y, _ = my_fn(1, 2, 3, 4)
x, y
```
