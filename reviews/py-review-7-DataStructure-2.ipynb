```python
my_basket = {'apple', 'orange', 'apple', 'pear', 'pear', 'orange', 'banana'}
my_basket # 어떤 원소가 있는지 보여준다.
```

```python
my_basket1 = {'apple', 'banana', 'watermelon'}
my_basket1
```

## Set의 연산

- 교집합: and(&)
- 합집합: or(|)
- 차집합: -
- 대칭차집합/XOR: ^

```python
my_basket
```

```python
my_basket1
```

```python
my_basket & my_basket1 # Intersection
```

```python
my_basket | my_basket1 # Union
```

```python
my_basket - my_basket1 # 차집합
```

```python
my_basket1 - my_basket
```

```python
my_basket ^ my_basket1 # == (my_basket - my_basket1) | (my_basket1 - my_basket): Symmetric Difference
```

## Set의 메서드

dir(set)으로 확인 가능하다.

- 교집합
- 합집합
- 차집합
- 원소추가
- 원소삭제
- set의 복제

```python
my_basket.intersection(my_basket1) # == my_basket & my_basket1
```

```python
my_basket.union(my_basket1) # == my_basket | my_basket1
```

```python
my_basket.difference(my_basket1) # == my_basket - my_basket1
```

```python
my_basket1.difference(my_basket) # == my_basket1 - my_basket
```

```python
my_basket.symmetric_difference(my_basket1) # == my_basket ^ my_basket1
```

```python
my_basket
```

```python
my_basket.add('grape') # 원소 추가
my_basket
```

```python
my_basket.remove('apple') # 원소 삭제
my_basket
```

```python
my_basket2 = my_basket.copy()
my_basket2
```

# 딕셔너리, 사전(dict)

## dict의 정의

- 빈 dict의 정의
- dict의 정의와 초기화

```python
my_dict = dict()
my_dict
```

```python
my_dict = {}
my_dict
```

```python
my_dict = dict({
    "a": 100,
    "b": 200,
    "c": 300,
    "d": 400
})
my_dict
```

```python
my_dict = {
    "a": 100,
    "b": 200,
    "c": 300,
    "d": 400
}
my_dict
```

```python
my_dict = dict(a=100, b=200, c=300, d=400)
my_dict
```

## dict의 메서드

- list: 값의 순서로 이루어진 자료구조
- dict: 키와 값으로 이루어진 자료구조 (리스트가 순서로 이루어진 자료구조인 것과 다름)
  - 값의 추가
  - 값의 변경
  - 값의 삭제

```python
my_dict = {
    "a": 100,
    "b": 200,
    "c": 300
}
my_dict
```

```python
my_dict["d"] = 400
my_dict
```

```python
my_dict["a"] = -100
my_dict
```

```python
del my_dict["d"]
my_dict
```

```python
my_dict["a"]
```

```python
# list에도 IndexError가 나오듯, dict에서 없는 key를 원하면 KeyError를 반환한다.
try:
    my_dict["d"]
except KeyError:
    None
```

```python
# 이것을 방지하기 위해 dict는 get 함수를 지원한다.
my_dict.get("a")
```

```python
print(my_dict.get("d"))
```

```python
my_dict.get("d", 0)
```

```python
my_dict.get("d", None) # == my_dict.get("d")
```

```python
print(my_dict.get("d", None))
```

```python
"a" in my_dict # 안티패턴
```

```python
"a" not in my_dict # 안티패턴
```

```python
"a" in my_dict.keys()
```

```python
"a" not in my_dict.keys()
```

```python
my_scores = {"a": 100, "b": 200}
my_scores
```

```python
"a" in my_scores.keys()
# "a" in ["a", "b", "c"]
# "a" in {"a": 100, "b": 200}
```

```python
100 in my_scores.values()
```

```python
my_scores.keys(), my_scores.values()
```
