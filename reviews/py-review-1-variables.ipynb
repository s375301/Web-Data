# 숫자

- 동적 타이핑 <-> 정적 타이핑
- 동적 타이핑 : javascript, python
- 정적 타이핑 : C/C++, Java

```python
# 변수의 값 저장
# 변수 - 정수
my_int = 3
my_int
```

```python
my_int + 1 # == 3 + 1
```

```python
# 변수 - 실수
my_float = 3.14
my_float
```

```python
my_float + 1
```

```python
my_float * 2
```

```python
# 정수를 나누어지지 않는 수로 나누면 실수가 된다.
my_int / 7
```

## type vs. isinstance

```python
# type : 변수의 데이터 타입을 알려주는 함수
type(my_int)
```

```python
type(my_float)
```

```python
# isinstance : 해당 데이터 타입이 맞는 지 확인하는 함수 (반환값 : True / False)
isinstance(my_int, int)
```

```python
isinstance(my_float, float)
```

## 불리언 (Boolean)

- 참(True) / 거짓(False)

```python
my_bool = True
my_bool
```

```python
# 다른 언어와 차이점은 True / False 의 첫 문자를 Capitalize 해야한다.
```

```python
# not : 반대의 값을 반환
not my_bool
```

```python
type(my_bool)
```

```python
isinstance(my_bool, bool)
```

```python
# 불리언은 bool 과 int 에서 둘다 True 이므로 type() 함수를 쓰는 것이 낫다.
isinstance(my_bool, int)
```

## 명시적 형변환

파이썬은 데이터에 대해 알아서 데이터 타입이 지정되는 동적 타이핑 기술을 따르며, 데이터 분석가가 필요할 시, 데이터 타입을 변환하는 것을 명시적 형변환이라고 부른다.
- int -> float
- str -> int / float
- str -> boolean

```python
my_float = 3.14
my_float
```

```python
type(my_float)
```

```python
# 명시적 형변환 (float -> int) : 소숫점 이하는 버려지고 정수인 부분만 취한다.
my_int = int(my_float)
my_int
```

```python
# 명시적 형변환 (int -> float) : 그 전 값의 소숫점 이하를 알 수 없기 때문에 3.0으로 소숫점만 표시한다.
float(my_int)
```

```python
type(my_int)
```

```python
# 명시적 형변환 (int -> str) : 따옴표가 생긴다.
str(my_int)
```

```python
my_str = 'True'
my_str
```

```python
# 명시적 형변환 (str -> bool)
bool(my_str)
```

```python
# 파이썬은 동적 타이핑을 지원한다.
# 동적 타이핑은 변수에 값을 넣을 때, 값의 데이터 타입이 결정 되는 것이다.
# 결정되는 타입은 크게 세 가지이다. (숫자/문자/불리언)
# 데이터 타입을 알기 위해서 type() 과 isinstance() 내장 함수를 이용한다.
```
