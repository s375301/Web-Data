```python
import requests
from bs4 import BeautifulSoup
```

```python
data = {
    "method": 'selectCnslWebPage',
    'menu_id': 'EMWPCnslWebInqL',
    'jndinm': 'EmwpCnslWebEJB',
    'methodnm': 'selectCnslWebPage',
    'context': 'NITS',
    # Note
    'pageIndex': 2,
}
```

```python
posts = my_soup.find('table', {'class': 'table'}).find_all('tr')
[post.find('td', {'class': 'td-num'}).text for post in posts if post.find('td', {'class': 'td-num'}) is not None]
```

```python
my_info = []

for post in posts:
    if post.find('td', {'class': 'td-num'}) is None:
        continue
    
    title = post.find('td', {'class': 'td-list'}).find('a').text.strip()
    namedate = post.find_all('td', {'class': 'td-date'})
    name = namedate[0].text
    date = namedate[1].text
    status = post.find('td', {'class': 'td-answer'}).text.strip()

    my_info.append({
        "title": title,
        "name": name,
        "date": date,
        "status": status,
    })

my_info
```

#### 3 Pages

```python
data = {
    "method": 'selectCnslWebPage',
    'menu_id': 'EMWPCnslWebInqL',
    'jndinm': 'EmwpCnslWebEJB',
    'methodnm': 'selectCnslWebPage',
    'context': 'NITS',
}

my_info = []
for page in range(1, 4):
    data.update({
        'pageIndex': page
    })
    res = requests.post(url, data=data)
    my_soup = BeautifulSoup(res.text, 'html.parser')
    posts = my_soup.find('table', {'class': 'table'}).find_all('tr')
    
    for post in posts:
        if post.find('td', {'class': 'td-num'}) is None:
            continue

        title = post.find('td', {'class': 'td-list'}).find('a').text.strip()
        namedate = post.find_all('td', {'class': 'td-date'})
        name = namedate[0].text
        date = namedate[1].text
        status = post.find('td', {'class': 'td-answer'}).text.strip()

        my_info.append({
            "title": title,
            "name": name,
            "date": date,
            "status": status,
        })

my_info
```
