```python
from fake_useragent import UserAgent
ua = UserAgent()
```

```python
url = "https://news.naver.com/"
res = requests.get(url, headers={'User-Agent': ua.random})
res
```

```python
res.text
```

```python
my_soup = BeautifulSoup(res.text, 'html.parser')
my_soup
```

```python
my_soup.find_all('div', {'class':'mtype_list_wide'})
```

```python
article_lists = my_soup.find_all('div', {'class':'mtype_list_wide'})
article_lists[-1]
```

```python
articles = article_lists[-1].find_all('a')
articles
```

```python
for article_elem in articles:
    print("CONTENT:", article_elem.find('strong').text)
    print("HYPERLINK:", article_elem['href'])
```
